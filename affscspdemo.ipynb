```python
# Create a discrete delta function to illustrate impulse responses
delta = pyscsp.discscsp.deltafcn(65,65)
```

```python
# For simplicity, we use smoothing with the sampled Gaussian kernel
affgauss = pyscsp.affscsp.samplaffgaussconv(delta, 8, 4, math.pi/6)
plt.imshow(-affgauss, cmap='gray', interpolation='nearest')
```

```python
# Directional derivative approximations are computed by applying
# small support derivative approximation masks to the smoothed
# image data
# Below is a first-order derivative in the direction pi/6
affgauss10 = pyscsp.affscsp.applyaffdirder(affgauss, 8, 4, math.pi/6, 1, 0)
plt.imshow(affgauss10, cmap='gray', interpolation='nearest')
```

```python
# Directional derivative approximations are computed by applying
# small support derivative approximation masks to the smoothed
# image data
# Below is a first-order derivative in the direction orthogonal to pi/6
affgauss01 = pyscsp.affscsp.applyaffdirder(affgauss, 8, 4, math.pi/6, 0, 1)
plt.imshow(affgauss01, cmap='gray', interpolation='nearest')
```

```python
# Directional derivative approximations are computed by applying
# small support derivative approximation masks to the smoothed
# image data
# Below is a second-order derivative in the direction pi/6
affgauss20 = pyscsp.affscsp.applyaffdirder(affgauss, 8, 4, math.pi/6, 2, 0)
plt.imshow(affgauss20, cmap='gray', interpolation='nearest')
```

```python
# Directional derivative approximations are computed by applying
# small support derivative approximation masks to the smoothed
# image data
# Below is a mixed second-order derivative based on the direction pi/6
affgauss11 = pyscsp.affscsp.applyaffdirder(affgauss, 8, 4, math.pi/6, 1, 1)
plt.imshow(affgauss11, cmap='gray', interpolation='nearest')
```
